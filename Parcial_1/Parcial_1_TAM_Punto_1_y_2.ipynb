```python
# Variables categóricas con nulos
cat_nulls = df.select_dtypes(include='object').isnull().sum()
cat_nulls = cat_nulls[cat_nulls > 0]

# Variables numéricas con nulos
num_nulls = df.select_dtypes(include=['int64', 'float64']).isnull().sum()
num_nulls = num_nulls[num_nulls > 0]

print("Variables categóricas con valores nulos:")
print(cat_nulls)

print("\nVariables numéricas con valores nulos:")
print(num_nulls)
```

```python
# Imputación categórica
for col in cat_nulls.index:
    df[col].fillna("None", inplace=True)

# Imputación numérica
for col in num_nulls.index:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Separar variables numéricas y categóricas
num_features = df.select_dtypes(include=['int64', 'float64']).drop('SalePrice', axis=1).columns.tolist()
cat_features = df.select_dtypes(include='object').columns.tolist()
```

```python
# Separar features y target
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
    ]
)
```

```python
from sklearn.model_selection import KFold

# Definir validación cruzada de 5 folds
cv = KFold(n_splits=5, shuffle=True, random_state=42)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np


# Definir función para MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100


```

```python
from sklearn.linear_model import LinearRegression, Lasso, ElasticNet, SGDRegressor, BayesianRidge
from sklearn.kernel_ridge import KernelRidge
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

# Diccionario con los modelos a comparar
models = {
    'Linear Regression': LinearRegression(),
    'Lasso': Lasso(random_state=42),
    'ElasticNet': ElasticNet(random_state=42),
    'Kernel Ridge': KernelRidge(),
    'SGD Regressor': SGDRegressor(random_state=42, max_iter=1000, tol=1e-3),
    'Bayesian Ridge': BayesianRidge(),
    'Gaussian Process': GaussianProcessRegressor(),
    'Random Forest': RandomForestRegressor(random_state=42),
    'SVR': SVR()
}

```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.pipeline import Pipeline

# Espacio de hiperparámetros para algunos modelos
param_grids = {
    'Lasso': {
        'model__alpha': [0.001, 0.01, 0.1, 1, 10]
    },
    'ElasticNet': {
        'model__alpha': [0.001, 0.01, 0.1, 1, 10],
        'model__l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9]
    },
    'Kernel Ridge': {
        'model__alpha': [0.1, 1, 10],
        'model__kernel': ['linear', 'rbf']
    },
    'Random Forest': {
        'model__n_estimators': [100, 200],
        'model__max_depth': [10, 20, None]
    },
    'SVR': {
        'model__C': [0.1, 1, 10],
        'model__kernel': ['linear', 'rbf']
    }
}



# Ejemplo de pipeline con preprocesador y modelo
pipe = Pipeline(steps=[('preprocessor', preprocessor),
                       ('model', Lasso(random_state=42))])

# Configurar GridSearchCV
grid_search = GridSearchCV(pipe, param_grid=param_grids['Lasso'],
                           cv=cv, scoring='neg_mean_absolute_error',
                           n_jobs=-1)

# Ejecutar búsqueda
grid_search.fit(X, y)

# Mejor hiperparámetro y score
print("Mejores hiperparámetros:", grid_search.best_params_)
print("Mejor MAE (negativo):", grid_search.best_score_)
```

```python
# RandomizedSearchCV ejemplo
random_search = RandomizedSearchCV(pipe, param_distributions=param_grids['Lasso'],
                                   n_iter=5, cv=cv, scoring='neg_mean_absolute_error',
                                   n_jobs=-1, random_state=42)

# Ejecutar búsqueda
random_search.fit(X, y)

# Resultados
print("Mejores hiperparámetros (Random):", random_search.best_params_)
print("Mejor MAE (negativo Random):", random_search.best_score_)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression, Lasso, ElasticNet, SGDRegressor, BayesianRidge
from sklearn.kernel_ridge import KernelRidge
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import KFold

# Assuming X, y, preprocessor, and cv are already defined as in your code

# Definir validación cruzada de 5 folds
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Lista para guardar los resultados
results = []

# Función para evaluar cada modelo
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def evaluar_modelo(nombre_modelo, modelo):
    # Create a simple transformer to convert sparse to dense
    class ToDenseTransformer():
        def fit(self, X, y=None):
            return self
        def transform(self, X):
            # Convierte la matriz dispersa a un array denso
            return X.toarray()

    # Crear pipeline con preprocesador, conversor a denso y modelo
    pipe = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('to_dense', ToDenseTransformer()), # Añade el paso de conversión a denso
        ('model', modelo)
    ])

    # Validación cruzada (predicciones)
    y_pred = cross_val_predict(pipe, X, y, cv=cv, n_jobs=-1)

    # Calcular métricas
    mae = mean_absolute_error(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    mape = mean_absolute_percentage_error(y, y_pred)

    # Guardar resultados
    results.append({
        'Modelo': nombre_modelo,
        'MAE': mae,
        'MSE': mse,
        'R2': r2,
        'MAPE (%)': mape
    })

    # Imprimir resultados
    print(f"Modelo: {nombre_modelo}")
    print(f"MAE: {mae:.2f}")
    print(f"MSE: {mse:.2f}")
    print(f"R²: {r2:.3f}")
    print(f"MAPE: {mape:.2f}%")
    print("-" * 40)


# Diccionario con los modelos a comparar
models = {
    'Linear Regression': LinearRegression(),
    'Lasso': Lasso(random_state=42),
    'ElasticNet': ElasticNet(random_state=42),
    'Kernel Ridge': KernelRidge(),
    'SGD Regressor': SGDRegressor(random_state=42, max_iter=1000, tol=1e-3),
    'Bayesian Ridge': BayesianRidge(),
    # Gaussian Process Regressor y SVR pueden ser computacionalmente caros y
    # también podrían tener problemas con datos dispersos o convergencia con
    # configuraciones predeterminadas y grandes conjuntos de datos. Los mantenemos
    # por ahora, pero tenlo en cuenta.
    'Gaussian Process': GaussianProcessRegressor(),
    'Random Forest': RandomForestRegressor(random_state=42),
    'SVR': SVR()
}

# Ejecutar la evaluación para cada modelo
for nombre, modelo in models.items():
    evaluar_modelo(nombre, modelo)
```

```python
# Convertir los resultados a DataFrame
resultados_df = pd.DataFrame(results).sort_values(by='MAE')
resultados_df.reset_index(drop=True, inplace=True)

# Mostrar tabla
resultados_df
```
